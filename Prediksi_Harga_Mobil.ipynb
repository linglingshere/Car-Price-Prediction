```python
data = data.drop(['CarName','fueltype','aspiration','doornumber','carbody','drivewheel','enginelocation','enginetype','cylindernumber','fuelsystem'], axis = 1)
sns.heatmap(data.corr())
```

```python
data = data.drop(['car_ID','symboling','carheight','stroke','compressionratio','peakrpm','citympg','highwaympg'],axis=1)
data.head()
```

FEATURE SELECTION
-> Setelah dianalisa kolom2 mana yang penting, kolom2 nggak penting gausah dipake

```python
feature_columns = ['wheelbase','carlength','carwidth','curbweight','enginesize','boreratio','horsepower']

X = data[feature_columns]
y = data['price']

print(X.head)
print(y.head)
```

SPLLITING DATA
- x: kolom independent
- y: kolom dependent

NOTE: test_size = 0.2 artinya 80% jadi data training, 20% jadi data test

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y)

print(X_train.head())
print(y_test.head())
```

SPECIFYING AND FITTING THE MODEL

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X_train, y_train)
```

EVALUATING THE MODEL

```python
# Membuat prediksi menggunakan data train
prediction = model.predict(X_train)
print('Prediksi Harga:', prediction[0])
print('Harga Asli:',y_train[:1])
```

```python
from sklearn.metrics import mean_absolute_error
# Membuat prediksi menggunakan data test
y_pred = model.predict(X_test)

# Menghitung MAE (Mean Absolute Error)
mae = mean_absolute_error(y_test, y_pred)
print('Mean Absolute Error:', mae)
print()
print(X_test.head())
print('Peringkat 5 teratas: ', y_pred[:5])

# Menghitung akurasi model (R-squared)
accuracy = model.score(X_test, y_test) * 100
print(f"{name} Accuracy: {accuracy:.2f}%")
```
